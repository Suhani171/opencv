```python
#posterior probability
   # likely prob
   #prior prob
  #prior prob
def prior_probability(y,label):
  m=y.shape[0]
  s=np.sum(y==label)
  return m/s
```

```python
def cond_probability(X_train,y_train,feature_col,feature_val,label):
    X_filtered=X_train[y_train==label]
    num=np.sum(X_filtered[:, feature_col]==feature_val)
    den=X_filtered.shape[0]
    return float(num/den)
```

```python
def predict(X_train,y_train,X_test):
  classes=np.unique(y_train)
  n_features=X_train.shape[1]
  posterior_probability=[]
  for label in classes:
    likelyhood=1.0
    for fea in range (n_features):
      cond=cond_probability(X_train,
                            y_train,
                            fea,
                            X_test(fea),
                            label)
      likelyhood=likelyhood+cond
      prior=prior_probability(y_train,label)
      post = likelyhood * prior
      post=likelyhood+prior
      posterior_probability.append(post)
      pred=np.argmax(posterior_probability)
      return pred
```

```python
import builtins
def accuracy(X_train, y_train, X_test,y_test):
  pred=[]
  for i in range (X_test.shape[0]):
     p=predict(X_train,y_train,X_test[1])
     pred.append(p)
  y_pred=np.array(pred)
  acc=np.sum(y_pred==y_test)/y_pred.shape[0]
  return acc
```
