```python
import keras_ocr
import matplotlib.pyplot as plt
from keras_ocr import tools
```

```python
pipeline=keras_ocr.pipeline.Pipeline()
```

```python
images=[
    keras_ocr.tools.read(img) for img in['image2.jpg',
                                          'image3.webp',
                                         'image1.png'

    ]
  ]
```

```python
len(images)
```

```python
plt.figure(figsize=(5,5))
plt.imshow(images[0])
```

```python
plt.figure(figsize=(5,5))
plt.imshow(images[1])
```

```python
plt.figure(figsize=(5,5))
plt.imshow(images[2])
```

```python
prediction_groups=pipeline.recognize(images)
```

```python
import matplotlib.pyplot as plt
import keras_ocr

# Assuming you have defined the variables `images` and `prediction_groups` before this code snippet

fig, axs = plt.subplots(nrows=len(images), figsize=(10, 10))

for ax, image, predictions in zip(axs, images, prediction_groups):
    keras_ocr.tools.drawAnnotations(image, predictions, ax=ax)

plt.show()
```
